```python
import pandas as pd
import os
from dotenv import load_dotenv
from tavily import TavilyClient
from typing import Any
from strands import Agent, tool
from strands.models.openai import OpenAIModel
 
load_dotenv()
 
```

```python
@tool
def search(query:str) -> dict [str,str]:
    client = TavilyClient(os.getenv("TAVILY_API_KEY"))
    response = client.search(
        query="Inteligencia Artificial y Talento Humano\nAgentes para Talento Humano",
        search_depth="advanced",
        max_results=10
    )
    return response

# query = "Inteligencia Artificial y Talento Humano\nAgentes para Talento Humano"
# result = search(query)
# result
```

```python
model = OpenAIModel(
    model_id=os.environ["LLM_DEPLOYMENT"],
    client_args={
        "api_key": os.environ["OPENAI_API_KEY"],
        "base_url": os.environ["LLM_API_URI"],
    },
    params={
        "max_tokens": 2048,
    }
)
```

```python
prompt = """
Eres un asistente de investigación  especializado en Inteligencia artificial aplicado al Talento Humano
Tu objetivos es investigar, analizar y sintetizar información actual y relevante sobre el uso de la Inteligencia Artificial en la gestión de Talento Humano, con un enfoque estratégico y práctico para organizaciones.
- Versionar el requerimiento del cliente en 3 preguntas enriquecidas
- Busqueda por cada pregunta enriquecida
- Recolectar la información y vas a seleccionar los que más se asemeje a la pregunta original del usuario
- Vas a responder al usuario con un json que contenga:
    [
        {
            "question_version": "the question versioned in previous steps",
            "URL": "URL coming from search tool",
            "score": "Te score coming from search tool",
        },... for all the selected search results
    ]
"""
agent = Agent(
    model=model,
    tools=[search],
    system_prompt=prompt
    
)
```

```python
query = "Inteligencia Artificial y Talento Humano\nAgentes para Talento Humano"
agent(query)
```
